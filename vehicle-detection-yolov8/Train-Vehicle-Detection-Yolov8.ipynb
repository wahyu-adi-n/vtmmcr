```python
from ultralytics import YOLO

model1 = YOLO('yolov8n')
model2 = YOLO('yolov8s')
model3 = YOLO('yolov8m')
model4 = YOLO('yolov8l')
model5 = YOLO('yolov8x')
```

```python
results1 = model1.train(
   data='/home/jovyan/catursupriyanto_dsn/vehicle-detection-yolov8/dataset/data.yaml',
   imgsz=640,
   epochs=50,
   batch=16,
   name='yolov8n_v8_yolovn'
)
```

```python
results2 = model2.train(
   data='/home/jovyan/catursupriyanto_dsn/vehicle-detection-yolov8/dataset/data.yaml',
   imgsz=640,
   epochs=50,
   batch=16,
   name='yolov8n_v8_yolovs'
)
```
